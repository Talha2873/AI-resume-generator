```python
import os
from openai import OpenAI
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# ---------- INPUT ----------
print("=== AI RESUME Generator ===\n")
resume_text = input("‚úçÔ∏è Paste your data here:\n")



# ---------- Extract Keywords ----------
keyword_prompt = f"""
Extract the skills/keywords from this data:

{resume_text}
"""
keywords = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are an ATS career assistant."},
        {"role": "user", "content": keyword_prompt}
    ]
).choices[0].message.content

# ---------- Generate resume  ----------
resume_prompt = f"""
*"Generate a one-page, modern, and visually appealing resume for me. I am a Computer Science student skilled in Python, NumPy, Machine Learning, and AI. I have completed Andrew Ng‚Äôs Machine Learning course and built projects such as an AI chatbot, a portfolio website, and a payment API system.
Structure the resume with these sections in order:
Header (Name, Contact Info, LinkedIn, GitHub/Portfolio)
Professional Summary (2‚Äì3 sentences highlighting my strengths in AI, Python, and problem-solving)
You must make it looks attractctive and professional to read.\
You must use heading,subheading,bullets and other markdown feature to make it attractive to read.
you a4 style and must make headings bold and there is no need to use same style or size from headings and data
or dont use * sign with them it must be look attractive and divide each sections by line"



Extracted Keywords:
{keywords}

resume:
"""
resume = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a professional career assistant."},
        {"role": "user", "content": resume_prompt}
    ]
).choices[0].message.content

# ---------- Resume Improvement Suggestions ----------
resume_feedback_prompt = f"""
Based on the following provided data and job ,experineces ,  make the resume
more ATS-friendly and better aligned with the job.

data:
{resume}


Provide suggestions in bullet points.
"""
suggestions = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful career advisor."},
        {"role": "user", "content": resume_feedback_prompt}
    ]
).choices[0].message.content

# ---------- OUTPUT ----------

print("\nüìÑ professional resume:\n", resume)
```
